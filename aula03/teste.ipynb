```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np 
```

```python
dados_x = np.linspace(0, 10, 10) # gera valores com a biblioteca numpy
dados_y = dados_x * dados_x # gera valores multiplicando um dado pelo outro
```

```python
plt.figure(figsize=(10,5), facecolor='gray')
Ax = plt.axes() # cria os eixos do grafico 
Ax.set_facecolor('white') # cria o fundo do grafico
plt.plot(dados_x, dados_y,
         color='red', linewidth=2,
         marker='o', linestyle = 'dashed', markeredgecolor='black')
```

```python
linhas = 2
colunas = 2
plt.figure(figsize=(12,12))
plt.subplot(linhas,colunas,1) # define o numero de linhas e colunas
plt.plot(dados_x,'r--')
plt.title("Grafico 1")
#grafico2
plt.subplot(linhas,colunas,2) # define o numero de linhas e colunas
plt.plot(dados_x * dados_x,'b-.')
plt.title("Grafico 2")
#grafico3
plt.subplot(linhas,colunas,3) # define o numero de linhas e colunas
plt.plot(dados_x/2,'c-.')
plt.title("Grafico 3")
#grafico4
plt.subplot(linhas,colunas,4) # define o numero de linhas e colunas
plt.plot(dados_y*2,'r--')
plt.title("Grafico 4")
plt.savefig("Exemplo subplot.jpg", dpi=300,transparent=True,
            bbox_inches= 'tight') # comando para salvar a figura
```

```python
#Grafico de barras
Figura = plt.figure() #Armazena na variavel Figura a figura
Grafico = Figura.add_axes([0,0,0.7,0.7]) # cria a grade do grafico 
Rotulos = ['A','B','C','D','E','F','G','H','I','J'] # cria os rotulos
Grafico.bar(Rotulos,dados_x,)
plt.title("Quantidade x Categoria")
plt.xlabel("Itens")
plt.ylabel("Quantidade")
```

```python
#Grafico 10%
Figura = plt.figure() #Armazena na variavel Figura a figura
Grafico = Figura.add_axes([0,0,0.7,0.7]) # cria a grade do grafico 
Rotulos = ['A','B','C','D','E','F','G','H','I','J'] # cria os rotulos
Grafico.bar(Rotulos,dados_x,color='blue')
Grafico.bar(Rotulos,dados_x*0.1,color='red')
plt.title("Quantidade x Categoria")
plt.xlabel("Itens")
plt.ylabel("Quantidade")
```

```python
#Grafico 10%
Figura = plt.figure() #Armazena na variavel Figura a figura
Grafico = Figura.add_axes([0,0,0.7,0.7]) # cria a grade do grafico 
Rotulos = ['A','B','C','D','E','F','G','H','I','J'] # cria os rotulos
Grafico.bar(Rotulos,dados_x,dados_x*0.1,color=['blue','red'])

plt.title("Quantidade x Categoria")
plt.xlabel("Itens")
plt.ylabel("Quantidade")
```
