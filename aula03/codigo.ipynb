```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('base de dados/monitoramento_tempo.csv') # realiza importação da base de dados
```

```python
df.head() # exibe as 4 primeiras linhas do dataframe
```

```python
# importa a biblioteca datetime
import datetime 
```

```python
df['data'] = pd.to_datetime(df['data']) # converte a coluna data
# de objeto 64 para datetime 
```

```python
df.info()
```

```python
plt.plot(df['data'],df['temperatura'])
plt.title("temperatura x ano",fontsize=15)
plt.xlabel("ano") # titulo do eixo x
plt.ylabel("temperatura") # titulo do eixo y
plt.legend("temperatura",loc='best') # coloca legenda no grafico
```
