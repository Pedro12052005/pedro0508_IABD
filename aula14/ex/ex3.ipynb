```python
# Importando as bibliotecas necessárias
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Definindo os dados dos setores e o número de produtos fabricados
setores = [1, 2, 3, 4, 5, 6, 7, 8]
produtos_fabricados = [100, 50, 15, 200, 500, 1000, 375, 450]

# Criando o DataFrame com os dados
df_setores = pd.DataFrame({
    'Setor': setores,
    'Produtos Fabricados': produtos_fabricados
})

# Aplicando o algoritmo K-means com 3 clusters
kmeans = KMeans(n_clusters=3, random_state=0)
df_setores['Cluster'] = kmeans.fit_predict(df_setores[['Produtos Fabricados']])

# Visualizando os clusters e os centros dos clusters
plt.figure(figsize=(10, 6))
plt.scatter(df_setores['Setor'], df_setores['Produtos Fabricados'], c=df_setores['Cluster'], s=100, cmap='viridis')

# Exibindo os centros dos clusters
centros = kmeans.cluster_centers_
plt.scatter(np.arange(1, 9), centros[kmeans.labels_, 0], c='red', s=200, marker='X', label='Centros dos Clusters')

# Configurando o gráfico
plt.title('Clusterização dos Setores com base na Produção')
plt.xlabel('Setor')
plt.ylabel('Produtos Fabricados')
plt.legend()
plt.grid(True)
plt.show()

# Exibindo o DataFrame final
print(df_setores)
```
