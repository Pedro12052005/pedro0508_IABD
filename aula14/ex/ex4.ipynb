```python
# Importando as bibliotecas necessárias
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Definindo os dados dos dias da semana e o número de clientes
dia_semana = [1, 2, 3, 4, 5, 6, 7]
num_clientes = [15, 20, 30, 100, 350, 500, 700]

# Criando o DataFrame com os dados
df_movimento = pd.DataFrame({
    'Dia da Semana': dia_semana,
    'Número de Clientes': num_clientes
})

# Aplicando o algoritmo K-means com 3 clusters para agrupar os dias de maior movimento
kmeans_movimento = KMeans(n_clusters=3, random_state=0)
df_movimento['Cluster'] = kmeans_movimento.fit_predict(df_movimento[['Número de Clientes']])

# Visualizando os clusters e os centros dos clusters
plt.figure(figsize=(10, 6))
plt.scatter(df_movimento['Dia da Semana'], df_movimento['Número de Clientes'], c=df_movimento['Cluster'], s=100, cmap='viridis')

# Exibindo os centros dos clusters
centros = kmeans_movimento.cluster_centers_
plt.scatter(np.arange(1, 8), centros[kmeans_movimento.labels_, 0], c='red', s=200, marker='X', label='Centros dos Clusters')

# Configurando o gráfico
plt.title('Clusterização dos Dias de Maior Movimento')
plt.xlabel('Dia da Semana')
plt.ylabel('Número de Clientes')
plt.legend()
plt.grid(True)
plt.show()

# Exibindo o DataFrame final com os clusters
df_movimento[['Dia da Semana', 'Número de Clientes', 'Cluster']]
```
