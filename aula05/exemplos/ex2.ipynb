```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Dados fornecidos
Q = np.array([100, 200, 300, 400, 500, 600, 700]).reshape(-1, 1)  # Quantidade de calor fornecida
T = np.array([30, 35, 40, 45, 60, 100, 150])  # Temperatura correspondente
```

```python
# Criação do modelo de regressão linear
model = LinearRegression()
model.fit(Q, T)
```

```python
# Previsão
def prever_temperatura(calor_fornecido):
    temperatura_prevista = model.predict(np.array([[calor_fornecido]]))
    return temperatura_prevista[0]
```

```python
# Exemplo de previsão
calor_fornecido = 550
temperatura = prever_temperatura(calor_fornecido)
print(f"Temperatura prevista para {calor_fornecido} unidades de calor: {temperatura:.2f}°C")
```

```python
# Plotar os dados e a linha de regressão
plt.scatter(Q, T, color='blue', label='Dados reais')
plt.plot(Q, model.predict(Q), color='red', label='Linha de regressão')
plt.xlabel('Quantidade de calor fornecida')
plt.ylabel('Temperatura (°C)')
plt.title('Regressão Linear: Calor vs Temperatura')
plt.legend()
plt.grid(True)
plt.show()
```
