```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
# Dados fornecidos
data = {
    'Velocidade de Operação': [10, 5, 8, 6, 12, 4],
    'Complexidade da Tarefa': ['Baixa', 'Alta', 'Média', 'Alta', 'Baixa', 'Alta'],
    'Manutenção Necessária': ['Baixa', 'Alta', 'Média', 'Alta', 'Baixa', 'Média'],
    'Classificação': ['Montagem', 'Teste', 'Montagem', 'Teste', 'Montagem', 'Teste']
}
```

```python
# Criar DataFrame
df = pd.DataFrame(data)
```

```python
# Codificação das variáveis categóricas
le_complexidade = LabelEncoder()
le_manutencao = LabelEncoder()
le_classificacao = LabelEncoder()
```

```python
df['Complexidade da Tarefa'] = le_complexidade.fit_transform(df['Complexidade da Tarefa'])
df['Manutenção Necessária'] = le_manutencao.fit_transform(df['Manutenção Necessária'])
df['Classificação'] = le_classificacao.fit_transform(df['Classificação'])
```

```python
# Separar features e target
X = df[['Velocidade de Operação', 'Complexidade da Tarefa', 'Manutenção Necessária']]
y = df['Classificação']
```

```python
# Criar o classificador de árvore de decisão
clf = DecisionTreeClassifier()
clf.fit(X, y)
```

```python

# Função para classificar novas máquinas
def classificar_maquina(velocidade, complexidade, manutencao):
    complexidade_encoded = le_complexidade.transform([complexidade])[0]
    manutencao_encoded = le_manutencao.transform([manutencao])[0]
    predicao = clf.predict([[velocidade, complexidade_encoded, manutencao_encoded]])
    return le_classificacao.inverse_transform(predicao)[0]

# Exemplos de classificação
print(classificar_maquina(11, 'Baixa', 'Baixa'))  # Deve retornar 'Montagem'
print(classificar_maquina(5, 'Alta', 'Alta'))      # Deve retornar 'Teste'
print(classificar_maquina(7, 'Média', 'Alta'))     # Deve retornar 'Teste'
```
